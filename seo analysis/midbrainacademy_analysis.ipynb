```python
final_df = temp_df[temp_df['Content'] == 'text/html']
```

```python
final_df = final_df[final_df['Status'] != 'Not Found']
```

```python
final_df.head(6)
```

```python
universal_keywords = []
for i in final_df['Meta Keyword 1']:
    #print(i)
    universal_keywords.extend([j.strip().lower() for j in str(i).split(',') if j != 'nan'])
```

```python
universal_unique_keywords = list(set(universal_keywords))
universal_unique_keywords.remove('nan')
```

```python
print(universal_unique_keywords)
```

```python
len(universal_unique_keywords)
```

```python
final_df.shape
```

here out of 182 number of links, 137 are unique keywords

```python
keyword_occur = {}
for i in universal_unique_keywords:
    keyword_occur[i] = universal_keywords.count(i)
```

```python
keyword_occur_descending = {k: v for k, v in sorted(keyword_occur.items(), reverse = True, key=lambda item: item[1])}
print(keyword_occur_descending)
```

```python
count = 0
pict_dict = {}
for i, j in keyword_occur_descending.items():
    if count == 40:
        break
    else:
        pict_dict[i] = j
    count += 1
```

```python
plt.bar(pict_dict.keys(),pict_dict.values())
plt.xticks(rotation=90)
plt.title("mostly reoccured Keywords in MIDBRAIN ACADEMY in their pages")
plt.show()
```

now finding the percentage of these most occur keywords in each link description

```python
most_occur_keywords = list(pict_dict.keys())
print(most_occur_keywords)
```

```python
each_description_keywords = {}
for i,j in zip(final_df['Meta Description 1'].index,final_df['Meta Description 1']):
    if str(j) != 'nan':
        count = 0
        for k in most_occur_keywords:
            count += j.lower().count(k.lower())
        #print(count,j)
        if count>2:
            each_description_keywords[i] = count
```
