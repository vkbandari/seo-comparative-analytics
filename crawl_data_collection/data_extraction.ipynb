```python
from bs4 import BeautifulSoup
import requests
```

```python
def scrape(link):
    response = requests.get(link)
    return(BeautifulSoup(response.text, 'html' ))
```

```python
import csv
```

```python
list_links = ["https://byjus.com"]
file = open('data.csv', 'w')
writer = csv.writer(file)

row_heading = ['link', 'entire_text', 'meta', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p']
writer.writerow(row_heading)


def find_links(link):
    sp = scrape(link)
    for i in sp.find_all("a"):
        temp_link = i.get('href')
        if temp_link == None or temp_link== '':
            continue
        elif temp_link[0] == '/':
            temp_link = link+temp_link
            #print(temp_link)
            if temp_link in list_links:
                continue
            else:
                list_links.append(temp_link)
                entire_text = ' '.join(sp.text.replace('\n', ' ').split())
                meta_xml = sp.find_all('meta')
                h1 = [' '.join(i.text.replace('\n','').split()) for i in sp.find_all('h1')]
                h2 = [' '.join(i.text.replace('\n','').split()) for i in sp.find_all('h2')]
                h3 = [' '.join(i.text.replace('\n','').split()) for i in sp.find_all('h3')]
                h4 = [' '.join(i.text.replace('\n','').split()) for i in sp.find_all('h4')]
                h5 = [' '.join(i.text.replace('\n','').split()) for i in sp.find_all('h5')]
                h6 = [' '.join(i.text.replace('\n','').split()) for i in sp.find_all('h6')]
                p  = [' '.join(i.text.replace('\n','').split()) for i in sp.find_all('p')]
                
                writer.writerow([temp_link, entire_text, meta_xml, h1,h2,h3,h4,h5,h6,p])
                
```

```python
for i,j in enumerate(list_links):
    if i <200:
        find_links(j)
        #print(j)
    else:
        break
```

```python
file.close()
```
